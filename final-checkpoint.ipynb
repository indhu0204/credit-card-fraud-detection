### now when i look into the performance i get to know that there are lot of variance in each perfomance so in order to treat such types we use either over sampling or under sampling to deal with majority and minority of the class and choose whioch model performs better 

```python
### starting with under sampling 
```

```python
genuine = df[df["Class"]==0]
fraud = df[df["Class"]==1]
```

```python
genuine.shape
```

```python
fraud.shape
```
